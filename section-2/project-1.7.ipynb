```python
import os 
from openai import OpenAI 
from dotenv import load_dotenv
import gradio as gr

load_dotenv(override=True)

api_key = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=api_key)
```

```python
## Prompt Engineering Functions

def travel_poster_prompt(destination, season, theme):
    return f""" 
Create a premium cinematic travel poster.

Destination: {destination}
Season: {season}
Theme: {theme}

Requirements
- Ultra realistic
- vibrant colors
- Tourist attractions
- professional travel agency poster
- beautiful travel agency poster
- beautiful typography saying 
"Visit {destination}" 
- High Quality
- Modern Travel Advertisement
"""

def boarding_pass_prompt(passenger, origin, destination, airline, flight, seat):
    return f""" 
Generate a realistic airline boarding pass.

Passenger Name: {passenger}

Airline: {airline}

Flight Number: {flight}

From: {origin}

To: {destination}

Seat: {seat}

Requirements
- Realistic Boarding Pass
- White and blue Airline theme
- QR code placeholder
- Professional layout
- Modern airline design
"""
```

```python
## dall-e travel poster generator
def generate_travel_poster(destination, season, theme):
    prompt = travel_poster_prompt(
        destination, season, theme
    )

    response = client.images.generate(
        model="gpt-image-1",
        prompt=prompt,
        size="1024x1024"
    )

    image_base64 = response.data[0].b64_json
    import base64 
    from io import BytesIO 
    from PIL import Image

    image = Image.open(
        BytesIO(base64.b64decode(image_base64))
    )

    return image
```

```python
## boarding pass generator
def generate_boarding_pass(
        passenger,
        origin,
        destination,
        airline,
        flight,
        seat,
):
    prompt = boarding_pass_prompt(passenger, 
                                  origin,
                                  destination,
                                  airline,
                                  flight,
                                  seat
                                  )
    
    response = client.images.generate(
        model="gpt-image-1",
        prompt=prompt,
        size="1024x1024"
    )

    image_base64 = response.data[0].b64_json

    import base64 
    from io import BytesIO
    from PIL import Image 

    image = Image.open(BytesIO(base64.b64decode(image_base64)))

    return image
```

```python
# ---------------------------------------
# Complete Workflow
# ---------------------------------------

def travel_generator(
    passenger,
    origin,
    destination,
    airline,
    flight,
    seat,
    season,
    theme,
):

    poster = generate_travel_poster(
        destination,
        season,
        theme,
    )

    boarding_pass = generate_boarding_pass(
        passenger,
        origin,
        destination,
        airline,
        flight,
        seat,
    )

    return poster, boarding_pass
```

```python
# ---------------------------------------
# Complete Workflow
# ---------------------------------------

def travel_generator(
    passenger,
    origin,
    destination,
    airline,
    flight,
    seat,
    season,
    theme,
):

    poster = generate_travel_poster(
        destination,
        season,
        theme,
    )

    boarding_pass = generate_boarding_pass(
        passenger,
        origin,
        destination,
        airline,
        flight,
        seat,
    )

    return poster, boarding_pass
```

```python
with gr.Blocks(title="AI Travel Generator") as demo:

    gr.Markdown("# ✈️ AI Travel Poster & Boarding Pass Generator")

    with gr.Row():

        passenger = gr.Textbox(label="Passenger Name")

        airline = gr.Textbox(
            label="Airline",
            value="OpenAI Airways"
        )

    with gr.Row():

        origin = gr.Textbox(label="Origin")

        destination = gr.Textbox(label="Destination")

    with gr.Row():

        flight = gr.Textbox(
            label="Flight Number",
            value="OA101"
        )

        seat = gr.Textbox(
            label="Seat",
            value="12A"
        )

    season = gr.Dropdown(
        ["Summer","Winter","Spring","Autumn"],
        value="Summer",
        label="Season"
    )

    theme = gr.Dropdown(
        [
            "Luxury",
            "Adventure",
            "Beach",
            "Mountains",
            "Romantic",
            "Nature",
            "Cultural"
        ],
        value="Luxury",
        label="Theme"
    )

    generate = gr.Button("Generate Images")

    with gr.Row():

        poster = gr.Image(
            label="Travel Poster"
        )

        boarding = gr.Image(
            label="Boarding Pass"
        )

    generate.click(
        fn=travel_generator,
        inputs=[
            passenger,
            origin,
            destination,
            airline,
            flight,
            seat,
            season,
            theme,
        ],
        outputs=[
            poster,
            boarding,
        ],
    )

demo.launch()
```
