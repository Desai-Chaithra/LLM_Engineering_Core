```python
# # Section 3 – Building a Multi-Model Airline Assistant
# Instead of using one model for everything, we'll let our application choose the appropriate model.
```

```python
from openai import OpenAI
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
# create a model router
def choose_model(user_question):
    question = user_question.lower()

    if "plan" in question:
        return "gpt-4.1"
    elif "book" in question:
        return "gpt-4o"
    else:
        return "gpt-4o-mini"
```

```python
def airline_assistant(user_question):
    model = choose_model(user_question)

    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "user",
                "content": user_question
            }
        ]
    )
    print(f"MODEL USED: {model}\n")
    return response.choices[0].message.content
```

```python
question="What is the baggage allowance?"
answer = airline_assistant(question)
print(answer)
```

##### Section 4 – System Prompts & Multi-shot Prompting
- Concepts Covered
- System Prompts
- Multi-shot (Few-shot) Prompting

```python
## imports
from openai import OpenAI
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
## create the system prompt
system_prompt = """ 
You are SkyFly Airline Assistant.

Your responsibilities are:
- Help passengers with airline-related questions.
- Be polite and professional.
- Give short and clear answers.
- Use Markdown formatting.
- If you don't know something, say you don't know.
- Never make up booking IDs or dlight details.
"""
```

```python
## ask a question using the prompt
question="What is the baggage allowance for international flights?"

response=client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {
            "role":"system",
            "content": system_prompt
        },
        {
            "role":"user",
            "content": question 
        }
    ]
)

print(response.choices[0].message.content)
```

##### What if we remove the System Prompt?
###### The model will still answer, but it no longer has instructions to behave like SkyFly Airline Assistant.

- instead of only giving instructions, we also provide examples.
- the model learns examples before answering the user's qestion.

## Comparison: System Prompt vs. Multi-Shot Prompting

| Feature | System Prompt | Multi-Shot Prompting |
| :--- | :--- | :--- |
| **Core Purpose** | Defines the assistant's overall role, persona, and strict rules. | Provides concrete example interactions to guide the model. |
| **Control Mechanism** | Controls foundational behavior, tone, and guardrails. | Demonstrates the desired response style, format, and pattern. |
| **Usage** | Usually written once at the beginning of the configuration. | Includes one or more example conversations built into the prompt history. |
| **Example** | *"You are an airline assistant. Always be polite and concise."* | **User:** `[Input A]` $\rightarrow$ **Assistant:** `[Output A]` <br>**User:** `[Input B]` $\rightarrow$ **Assistant:** `[Output B]` |

> **Pro Tip:** For the best results, use both together! Use a **System Prompt** to lock in the absolute rules, and **Multi-Shot examples** to show the AI exactly what a "perfect" answer looks like.

```python
import gradio as gr 
from openai import OpenAI
```

```python
client = OpenAI(
    api_key = os.getenv("OPENAI_API_KEY")
)
```

```python
system_prompt = """ 
You are SkyFly Airline Assistant.
Help passengers with airline-related questions.

Be polite.
Use Markdown.
Keep answers short.
"""
```

```python
## create the call back function
# this is the most important concept in this section

def airline_assistant(question):
    response=client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role":"system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": question
            }
        ]
    )
    return response.choices[0].message.content
```

##### What is a Callback Function?
- Gradio doesn't know how to answer airline questions.
- It only knows: when the user clicks submit, call this python function. that python function is called a callback.

```python
question="What is the baggage allowance for international flights?"

airline_assistant(question)
```

```python
# creating a gradio interface
app=gr.Interface(
    fn=airline_assistant,
    inputs=gr.Textbox(
        label="Ask Your Airline Question"
    ),
    outputs = gr.Markdown(
        label="SkyFly Response"
    ),
    title="✈️ SkyFly Airline Assistant",
    description="Ask questions about flights, baggage, bookings, and airline policies."
)
```

```python
app.launch(share=True)
""" if share=True means, the url will not be a local host
rather it will create a temporary public URL, allowing 
someone else to access your app over the internet.
"""
```

##### Section 6 – Markdown Responses & Streaming with OpenAI + Gradio

- Markdown responses 
- streaming with OpenAI API
- Building a Streaming ChatBot with Gradio

```python
import gradio as gr 
from openai import OpenAI
```

```python
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
```

```python
system_prompt = """
You are SkyFly Airline Assistant.

Help passengers with airline-related questions.

Be polite.

Use Markdown.

Keep answers short and clear.
"""
```

```python
def airline_assistant(question):

    stream = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": question
            }
        ],
        stream=True
    )

    answer = ""

    for chunk in stream:

        content = chunk.choices[0].delta.content

        if content:
            answer += content
            yield answer
```

```python
# build the interface
app = gr.Interface(
    fn=airline_assistant,

    inputs=gr.Textbox(
        label="Ask Your Airline Question"
    ),
    outputs=gr.Markdown(
        label="SkyFly Response"
    ),
    title="✈️ SkyFly Airline Assistant",
    description="Streaming Airline Assistant powered by OpenAI."

)
```

```python
app.launch(share=True)
```

##### Section 7 – Building a ChatGPT-like Airline Assistant

```python
import gradio as gr

def airline_chat(message, history):

    # Conversation starts with the System Prompt
    messages = [
        {
            "role": "system",
            "content": system_prompt
        }
    ]

    # Add previous conversation
    messages.extend(history)

    # Add the latest user message
    messages.append(
        {
            "role": "user",
            "content": message
        }
    )

    # Send conversation to OpenAI
    stream = client.chat.completions.create(
        model="gpt-4o",
        messages=messages,
        stream=True
    )

    answer = ""

    # Stream the response
    for chunk in stream:

        content = chunk.choices[0].delta.content

        if content:
            answer += content
            yield answer


gr.ChatInterface(
    fn=airline_chat,
    title="✈️ SkyFly Airline Assistant",
    # type="messages"
).launch(
    share=True,
    auth=("admin", "airline123")
)
```
