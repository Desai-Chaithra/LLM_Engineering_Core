1. Choosing the Right LLM Model — Selection Strategies
What is Model Selection?
Choosing an LLM means selecting the model that provides the best tradeoff between
Intelligence
Speed
Cost
Context Window
Latency
Deployment Constraints

| Factor         | Meaning                   | Example                  |
| -------------- | ------------------------- | ------------------------ |
| Accuracy       | Reasoning capability      | GPT-5.5 > smaller models |
| Latency        | Response speed            | GPT-5.5 nano             |
| Cost           | Price per token           | Smaller models cheaper   |
| Context Window | Maximum input length      | 128K+ models             |
| Tool Calling   | Supports function calling | GPT family               |
| Multimodal     | Image + Audio + Text      | GPT-5.5                  |
| Open Source    | Self-hosting              | Llama                    |
| Fine-tuning    | Custom training           | Open-source models       |



- if a smaller model achieves similar quality, use the smaller models.
- use large models for complex reasoning like coding assistant, medical qa, legal analysis etc.
- use context models, hybrid pipeline.
- the goal is not about highest intelligence, it is about within budget.

## Chinchilla scaling law
- before chinchilla - researchers believed that,
bigger model = better model
so they increased parameters but not the training data
- performace depends on model parameters and training tokens

## understanding ai benchmarks
- benchmarks evaluate llm capabilities objectively
- they compare models using same tasks.

## GPQA - graduate level google prood q&a
- tests physics, biology, chemistry. questions are intentionally difficult to answer by simple web search

## MMLU-Pro
- extension of mmlu.
- it tests medicine, history, mathematics, computer science, economics, law over many professional-level subjects.

## HLE - humanity's last exam
- one of the hardest public reasoning benchmarks. designed so that even frontier models struggle.
- it measures deep reasoning, multi-step thinking, scientific understanding, mathematical ability.

## limitations of ai benchmarks
1. data contamination - the benchmark accidentally appears in the model's training data.
2. overfitting - companies optimize models specifically for benchmark datasets.
3. narrow-evaluation - benchmarks cannot fully evaluate these.
4. static dataset - questions stay the same, models eventually memorize them.
---


## building a connect four leaderboard(reasoning benchmark)
- a fun way to compare reasoning models.
- it measures win rate, illegal moves, average thinking time, strategic quality.

## navigating ai leaderboards
- leaderboards compare models using standardized evaluations.
they help engineers choose suitable models.
1. aritificial analysis: provides independent evaluations across many frontier models.
- they track intelligence, cost, coding, speed, context, latency
---
2. hugging face open llm leaderboard: ranks open-source models. they measure reasoning, knowledge, math, coding, general capability.
---
3. lymsys chatbot arena: models compete in anonymous head-to-head conversations.
humans vote for the better response.
4. live bench: continuously updated with fresh questions to reduce benchmark contamination and evalute newer capabilities.
